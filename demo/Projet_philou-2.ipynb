```python
test = pd.read_excel('/Users/remi/Downloads/Classeurbo_inflation.xlsx')
test = test[['rang','titre','distributeur','date_sortie','pays','budmeuros','budmdollars','totalentrees']]
test
```

```python
test['annee_de_sortie'] = pd.to_datetime(test['date_sortie']).dt.year
```

```python
test['mois_de_sortie'] = pd.to_datetime(test['date_sortie']).dt.month
```

```python
test
```

```python
l=[]

list_pays = list(test['pays'])
for n in list_pays:
    if str(n).startswith('FRA'):
        n = 'FRA'
        l.append(n)
    elif str(n).startswith('USA'): 
        n = 'USA'
        l.append(n)
    else:
        n='Autre'
        l.append(n)
 

len(l)

test['pays'] = l 
```

```python
test
```

```python
a =[]

for n in test['budmeuros']:
    if math.isnan(n) == True:
        a.append(0)
    else:
        a.append(n)
```

```python
b =[]

for n in test['budmdollars']:
    if math.isnan(n) == True:
        b.append(0)
    else:
        b.append(n)

```

```python
test['budmeuros'] = a
test['budmdollars'] = b
test['budget'] = test['budmeuros'] + test['budmdollars']
```

```python
test.drop(columns=['budmeuros','budmdollars'],inplace=True)
```

```python
l=[]

list_mois = list(test['mois_de_sortie'])


for n in list_mois:
    if n<=3:
        n = 'Hiver'
        l.append(n)
    elif n>=3 and n<=6: 
        n = 'Printemps'
        l.append(n)
    elif  n>=6 and n<=9:
        n='Ete'
        l.append(n)
    else:
        n = 'Automne'
        l.append(n)
 

len(l)

```
